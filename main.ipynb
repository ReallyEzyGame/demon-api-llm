# Imoprt Libraries

```python
import subprocess
import sys
import time
```

# Build Model & Server + Init Server
The code to build model & server is inside /source/api.py


```python
# for writing on the log file
log_file = open("server.log", "w")


# for running on different cell
process = subprocess.Popen(
    [sys.executable, "./source/api.py"],
    stdout=log_file,
    stderr=subprocess.STDOUT,
    start_new_session=True  # prevent Colab from dismiss the program
)

print(f"Tiến trình Server đang chạy ngầm (PID: {process.pid})")
print("Đang tải Model vào GPU, quá trình này mất khoảng 1-2 phút...\n")
print("-" * 50)

# automatically reading file and show the proccess
with open("server.log", "r") as f:
    while True:
        line = f.readline()
        if not line:
            time.sleep(0.5)
            continue

        # log to screen showing the proccess
        print(line, end="")

        # Server is ready
        if "Application startup complete" in line:
            print("-" * 50)
            print("SERVER ĐÃ SẴN SÀNG!")
            break
        # interupted
        if "Shutting down" in line:
            print("-" * 50)
            print("LỖI: Server đã bị tắt!")
            break
```
